```python
import numpy as np
```

```python
trainData = [
    #(x,y) pairs
    ((0,2),1),
    ((-2,0),1),
    ((1,-1),-1)
]
```

```python
def phi(x):
    return np.array(x)
```

```python
def initialWeightVector():
    return np.zeros(2)
```

```python
def hingeLoss(w):
    return (1/len(trainData)) * (sum(max(1-w.dot(phi(x))*y,0) for x,y in trainData))
```

```python
def gradientHingeLoss(w):
    return (1/len(trainData)) * sum(-phi(x)*y if 1-w.dot(phi(x))*y > 0 else 0 for x,y in trainData)
```

```python
def gradientDescent(func, gradientFunc, initialWeights):
    w = initialWeights()
    lr = 0.1
    for x in range(500):
        value = func(w)
        gradient = gradientFunc(w)
        w = w - lr * gradient
        print(f'epoch{x}: weights = {w}, F(weights): {value}, gradient: {gradient}')
```

```python
gradientDescent(hingeLoss, gradientHingeLoss, initialWeightVector)
```
