# This notebook explores the implementation of Gradient Descent in 3D(optimization of 2 parameters)

```python
# import the libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
# define a function to minimize
def z_func(x,y):
    return np.sin(5*x) * np.cos(5*y) / 5

# we minimize the function by optimizing for x and y
```

```python
# obtain the gradient
def gradient_z(x,y):
    deriv_x = np.cos(5*x) * np.cos(5*y)
    deriv_y = -1 * np.sin(5*x) * np.sin(5*y)
    return deriv_x, deriv_y
```

```python
# define range of x and y values for plotting
x = np.arange(-1,1,0.01)
y = np.arange(-1,1,0.01)

# create meshgrid
X,Y = np.meshgrid(x,y)
```

```python
# calculate Z values
Z = z_func(X,Y)
```

```python
# plot the function
ax = plt.subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap='coolwarm')
```

```python
# initialize a starting point
current_pos = [0, -0.5, z_func(0, -0.5)]
```

```python
# plot the graph with initialized parameters
ax = plt.subplot(projection='3d',computed_zorder=False)
ax.plot_surface(X,Y,Z,zorder=0,cmap='coolwarm')
ax.scatter3D(current_pos[0],current_pos[1],current_pos[2],zorder=1,color='r')
```

```python
# run the gradient descent algorithm

# set learning rate
learn_rate = 0.01
ax = plt.subplot(projection='3d',computed_zorder=False)
ax.plot_surface(X,Y,Z,zorder=0,cmap='coolwarm')

# iterate through loop
for _ in range(1000):
    x = current_pos[0]
    y = current_pos[1]
    z = current_pos[2]
    
    step_size_x = gradient_z(x,y)[0] * learn_rate
    current_pos[0] = current_pos[0] - step_size_x

    step_size_y = gradient_z(x,y)[1] * learn_rate
    current_pos[1] = current_pos[1] - step_size_y
ax.scatter3D(current_pos[0],current_pos[1],
            z_func(current_pos[0],current_pos[1]),
            color='r',zorder=1)  
```

We see that in the above plot, initialized point has reached minimum of the function. So this is the implementation of gradient descent in 3D.
