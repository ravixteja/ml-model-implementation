# In this notebook, we explore implementation of gradient descent on a simple 2D function (optimization of single parameter)

The Gradient Descent algorithm:<br>
1. Define a function to optimize
2. Take gradient (here derivative since we are optimizing a single parameter)
3. Initialize learning rate
4. Initialize parameter
5. Calculate Step Size
6. Update the parameter (new = old - step size)

Iterate from step 4 till max number of iterations is reached

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define a function to minimize
def y_func(x):
    return x ** 2
```

```python
# define derivative
def dydx(x):
    return 2 * x


# set learning rate
n = 0.1
```

```python
# take a range of values of x
x = np.arange(-100,100,0.1)
y = y_func(x)

# plot the function
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y as function of x")
```

```python
# initialize a starting point
current_position = [-95,y_func(-95)]
# plot the current position in the graph
plt.plot(x,y)
plt.scatter(current_position[0],current_position[1],color='r',label='Initial Parameter')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("y as function of x")
```

```python
# plot the function to visualize how the point moves towards minimum
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent in 2D")

# apply the gradient descent rule
# it runs in a loop
for _ in range(1000):

    # calculate step size
    step_size = dydx(current_position[0]) * n
    
    # update the position of point
    current_position[0] = current_position[0] - step_size
    current_position[1] = y_func(current_position[0])

    
    plt.scatter(current_position[0],current_position[1],color='r')
```

So the above graph shows how Gradient Descent Alogrithm works.<br>
- When the slope is steep, it takes large steps for parameters
- When the slope is shallow, it takes very small steps for parameters, since it approaches to minimum.
This is indicated by red dots.
